```python
# Instalação (caso esteja rodando localmente, pode ignorar no Colab)
!pip install pandas

# Importando bibliotecas
import pandas as pd

# Link da base de dados CSV (FUNCIONA)
url = "https://raw.githubusercontent.com/mariamills/World-Happiness/main/happiness_data_2023.csv"

# Carregando os dados
df = pd.read_csv(url)

# Renomeando colunas para português
df = df.rename(columns={
    'Country name': 'País',
    'Regional indicator': 'Região',
    'Ladder score': 'Pontuação',
    'Standard error of ladder score': 'Erro padrão da pontuação',
    'Logged GDP per capita': 'PIB per capita (log)',
    'Social support': 'Suporte social',
    'Healthy life expectancy': 'Expectativa de vida saudável',
    'Freedom to make life choices': 'Liberdade para fazer escolhas',
    'Generosity': 'Generosidade',
    'Perceptions of corruption': 'Percepção de corrupção',
    'Ladder score in Dystopia': 'Pontuação na distopia',
    'Explained by: Log GDP per capita': 'Explicado por: PIB per capita (log)',
    'Explained by: Social support': 'Explicado por: Suporte social',
    'Explained by: Healthy life expectancy': 'Explicado por: Expectativa de vida saudável',
    'Explained by: Freedom to make life choices': 'Explicado por: Liberdade para fazer escolhas',
    'Explained by: Generosity': 'Explicado por: Generosidade',
    'Explained by: Perceptions of corruption': 'Explicado por: Percepção de corrupção'
})

# Exibindo as primeiras 5 linhas
print("Visualização inicial dos dados (com colunas em português):")
display(df.head())

# Informações básicas sobre as colunas
print("\nInformações do DataFrame:")
df.info()

# Verificando valores ausentes
print("\nValores ausentes por coluna:")
print(df.isnull().sum())

# Estatísticas descritivas das colunas numéricas
print("\nEstatísticas descritivas:")
display(df.describe())

# Remover linhas com valores ausentes (se desejar)
df_clean = df.dropna()

# Verificando novamente se há valores ausentes
print("\nValores ausentes após limpeza:")
print(df_clean.isnull().sum())

# Salvando o CSV limpo
df_clean.to_csv("world_happiness_clean.csv", index=False)
print("\nCSV limpo salvo com sucesso como 'world_happiness_clean.csv'")
```
